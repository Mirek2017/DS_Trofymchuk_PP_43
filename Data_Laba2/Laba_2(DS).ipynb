```python
df1_filtered = df1[df1['SALE PRICE'] >= 1000]
# Топ-10 найдорожчих типів будівель 
top_expensive_buildings = df1_filtered.groupby('BUILDING CLASS CATEGORY')['SALE PRICE'].mean().nlargest(10).reset_index()
top_expensive_buildings.columns = ['BUILDING CLASS CATEGORY', 'AVERAGE PRICE']
top_expensive_buildings = top_expensive_buildings.applymap(lambda x: '{:,.2f}'.format(x) if isinstance(x, (int, float)) else x)
top_expensive_buildings
```

```python
# Топ-10 найдешевших типів будівель
top_cheap_buildings = df1_filtered.groupby('BUILDING CLASS CATEGORY')['SALE PRICE'].mean().nsmallest(10).reset_index()
top_cheap_buildings.columns = ['BUILDING CLASS CATEGORY', 'AVERAGE PRICE']
top_cheap_buildings = top_cheap_buildings.applymap(lambda x: '{:,.2f}'.format(x) if isinstance(x, (int, float)) else x)
top_cheap_buildings
```

```python
# Топ-10 типів будівель з відсутньою площею
missing_land_square_feet = df1[df1['GROSS SQUARE FEET'].isnull() & df1['GROSS SQUARE FEET'] == 0]
top_10_missing_land_types = missing_land_square_feet['BUILDING CLASS CATEGORY'].value_counts().nlargest(10).reset_index()
top_10_missing_land_types.columns = ['BUILDING CLASS CATEGORY', 'BUILDING WITHOUT SQUARE']
top_10_missing_land_types
```

```python
# Топ-10 найдорожчих типів будівель за вартістю одиниці площі
df2_filtered = df1[df1['SALE PRICE'] >= 1000]
df2_filtered = df2_filtered.dropna(subset=['GROSS SQUARE FEET'])
df2_filtered = df2_filtered[df2_filtered['GROSS SQUARE FEET'] != 0]
df2_filtered = df2_filtered[df2_filtered['LAND SQUARE FEET'] != 0]
df2_filtered['PRICE PER GROSS SQFT'] = df2_filtered['SALE PRICE'] / df2_filtered['GROSS SQUARE FEET']
df2_filtered['PRICE PER LAND SQFT'] = df2_filtered['SALE PRICE'] / df2_filtered['LAND SQUARE FEET']
top_exp = df2_filtered.groupby('BUILDING CLASS CATEGORY')['PRICE PER LAND SQFT'].mean().nlargest(10).reset_index()
top_exp.columns = ['BUILDING CLASS CATEGORY', 'PRICE PER LAND SQFT']
top_exp
#top_exp = df2_filtered.groupby('BUILDING CLASS CATEGORY').agg({
#    'PRICE PER GROSS SQFT': 'mean',
#    'PRICE PER LAND SQFT': 'mean' 
#}).nlargest(10, 'PRICE PER LAND SQFT')
#top_exp
```

```python
# Топ-5 найдешевших типів будівель за вартістю одиниці площі
top_cheap_buildings = df2_filtered.groupby('BUILDING CLASS CATEGORY')['PRICE PER LAND SQFT'].mean().nsmallest(10).reset_index()
top_cheap_buildings.columns = ['BUILDING CLASS CATEGORY', 'PRICE PER LAND SQFT']
top_cheap_buildings
```

```python
# Всі райони включають топ-3 підрайони за продажами і MIN & MAX PRICE, AVG PRICE, AVG PRICE PER SQFT для кожного підрайона
top_neighborhoods = df2_filtered.groupby(['NY BOROUGH', 'NEIGHBORHOOD'])['SALE PRICE'].count().groupby('NY BOROUGH', group_keys=False).nlargest(3).reset_index()
results = pd.DataFrame()
for index, row in top_neighborhoods.iterrows():
    borough, neighborhood = row['NY BOROUGH'], row['NEIGHBORHOOD']    
    subset = df2_filtered[(df2_filtered['NY BOROUGH'] == borough) & (df2_filtered['NEIGHBORHOOD'] == neighborhood)]    
    min_price = subset['SALE PRICE'].min()
    max_price = subset['SALE PRICE'].max()    
    avg_price = subset['SALE PRICE'].mean()
    avg_price_per_lsqft = subset['PRICE PER LAND SQFT'].mean()
    avg_price_per_gsqft = subset['PRICE PER GROSS SQFT'].mean()   
    results = pd.concat([results, pd.DataFrame({
        'NY BOROUGH': [borough],
        'NEIGHBORHOOD': [neighborhood],
        'MIN PRICE': [min_price],
        'MAX PRICE': [max_price],        
        'AVG PRICE': [avg_price],
        'AVG PRICE PER LAND SQFT': [avg_price_per_lsqft],
        'AVG PRICE PER GROSS SQFT': [avg_price_per_gsqft]
    })])
    results = results.applymap(lambda x: '{:,.2f}'.format(x) if isinstance(x, (int, float)) else x)
results
```

```python
grouped_by_borough = df2_filtered.groupby('NY BOROUGH')
result = grouped_by_borough.agg({
    'SALE PRICE': ['min', 'max', 'mean', 'median'],
    'PRICE PER LAND SQFT': 'mean',
    'PRICE PER GROSS SQFT': 'mean'
})
result.columns = ['MIN PRICE', 'MAX PRICE', 'AVG PRICE', 'MEDIAN PRICE', 'AVG PRICE PER LAND SQFT', 'AVG PRICE PER GROSS SQFT']
result = result.applymap(lambda x: '{:,.2f}'.format(x) if isinstance(x, (int, float)) else x)
result
```

```python
df6 = df1.dropna(subset=['YEAR BUILT'])
df6.loc[df6['YEAR BUILT'] == 220, 'YEAR BUILT'] = 2022
bins = [1800, 1900, 1925, 1950, 1975, 2000, 2023]
labels = ['1800-1900', '1900-1925', '1925-1950', '1950-1975', '1975-2000', '2000-2023']
df6['YEAR GROUP'] = pd.cut(df6['YEAR BUILT'], bins=bins, labels=labels, right=False)
plt.figure(figsize=(9, 5))
sns.countplot(y='YEAR GROUP', hue = 'YEAR GROUP', data=df6, order=df6['YEAR GROUP'].value_counts().index, palette='viridis')
plt.title('SALES BY YEAR BUILT')
plt.xlabel('SALES COUNT')
plt.ylabel('')
plt.show()
```

```python
# Топ-3 будівлі (збудовані між 1800 і 1900 роками) по продажам, їхня кількість, max, min і середня вартість.
df6 = df6[df6['SALE PRICE'] >= 1000]
filtered_df6 = df6[(df6['YEAR BUILT'] >= 1800) & (df6['YEAR BUILT'] <= 1900)]
top_buildings = filtered_df6.groupby(['NY BOROUGH', 'BUILDING CLASS CATEGORY'])['SALE PRICE'].count().groupby('NY BOROUGH', group_keys=False).nlargest(3).reset_index()
results = pd.DataFrame()
for index, row in top_buildings.iterrows():
    borough, building = row['NY BOROUGH'], row['BUILDING CLASS CATEGORY']    
    subset = filtered_df6[(filtered_df6['NY BOROUGH'] == borough) & (filtered_df6['BUILDING CLASS CATEGORY'] == building)]    
    min_price = subset['SALE PRICE'].min()
    max_price = subset['SALE PRICE'].max()    
    avg_price = subset['SALE PRICE'].mean()  
    count = subset['SALE PRICE'].count()  
    results = pd.concat([results, pd.DataFrame({
        'NY BOROUGH': [borough],
        'BUILDING CLASS CATEGORY': [building],
        'SALES COUNT': [count],
        'MIN PRICE': [min_price],
        'MAX PRICE': [max_price],        
        'AVG PRICE': [avg_price]        
    })])
    results = results.applymap(lambda x: '{:,.2f}'.format(x) if isinstance(x, (int, float)) else x)
results
```
